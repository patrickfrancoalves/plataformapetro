```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

# Procura nomes no ORCID

```python
# Reading the SAS Transport file
df_pesq = pd.read_sas('re_pesquisador_infraestrutura.sas7bdat',format = 'sas7bdat', encoding='latin-1')
nomes_pesquisadores = df_pesq.NOME_COMPLETO.unique()
```

```python
nomes_pesquisadores[5758:5760]
```

```python
for item in nomes_pesquisadores[5759:]:
    driver  = webdriver.Chrome()
    driver.get('https://orcid.org/orcid-search/search?searchQuery='+item)

    time.sleep(6)

    headlines = driver.find_elements(By.CLASS_NAME, "orcid-id-column")
    lista_orcid = []

    for headline in headlines:
        lista_orcid.append(headline.text)

    time.sleep(3)
    
    try:
        lista_orcid.remove('ORCID ID')
        df = pd.DataFrame({'orcid': lista_orcid[0:6]})
        df['nome'] = item
        df.to_csv('.\orcid\df_{0}.csv'.format(item), encoding="utf-8-sig")
    except:
        pass
    time.sleep(1)
    driver.close()
    
```
