```python
from pathlib import Path
import glob
import pandas as pd

# Get data file names
filenames = glob.glob("*.csv")
print(filenames)
```

```python
colunas = ["Num_Processo","Titulo_Por","Titulo_Ing","Beneficiario","Instituicao","Cidade_Inst","Inst_Parceira",
"Empresa","Municipio","Pesq_Responsavel","Pesq_Principais","Pesq_Associados","Supervisor","Local_Pesquisa",
"Pesq_Visitante","Inst_Pesq_Visitante","Modalidade_Apoio","Grande_Area","Area","Sub_Area","Assuntos",
"Data_Inicio","Data_Termino","Acordo_Cooper_FAPESP","Inst_Exterior","Pais_Inst_Convenio","Pesq_Resp_Exterior",
"Resumo_Por","Resumo_Ing","Processos_Vinc"]
```

```python
#on_bad_lines='skip'

dfs = []

for filename in filenames:
    dfs.append(pd.read_csv(filename, skiprows=1, names=colunas,  delimiter=';', header=None))

# Concatenate all data into one DataFrame
fapesp = pd.concat(dfs, ignore_index=True)
#cod_equip['fica'] = cod_equip['codigos'].str.slice(0, 3)
#cod_equip['site_mcti'] = "https://pnipe.mcti.gov.br/laboratory/" + cod_equip['lab'].astype(str)
```
