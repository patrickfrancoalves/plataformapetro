```python
#https://stackoverflow.com/questions/70764097/how-can-i-ping-a-specific-port-and-report-the-results
import socket
import random
import time
#print(random.randint(0,3))

def check(host,port,timeout=2):
    sock = socket.socket(socket.AF_INET,socket.SOCK_STREAM) #presumably 
    sock.settimeout(timeout)
    try:
       sock.connect((host,port))
    except:
       return False
    else:
       sock.close()
       return True

#print(check('10.221.2.9',5558,timeout=1))


#Check port open for range of ports
#https://pandeysudhendu.medium.com/using-python-to-check-tcp-socket-connectivity-8384a036833c
for i in range(50300,50599):
    time.sleep(random.randint(0,3))
    print(i,':',check('10.221.2.9',i,timeout=1))
    
```

```python
import socket
import time
import random

def check_port(port):
    import socket, errno
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    random.randint(0,3)
    try:        
        s.bind(("127.0.0.1", port))
    except socket.error as e:
        if e.errno == errno.EADDRINUSE:
            print("Port is already in use", port)
        else:
            print(e)
            print(f"****Port: {port} cannot be acessed****")
            return False
    s.close()
    return True

# Check port open for range of ports
for i in range (1000):
    random.randint(0,3)
    check_port(i)
    print()
```
