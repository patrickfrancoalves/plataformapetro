# Example: StoNED with CNLS [.ipynb]
- #https://pystoned.readthedocs.io/en/latest/examples/StoNED/mom.html


```python
# import packages
from pystoned import CNLS, StoNED
from pystoned.dataset import load_Finnish_electricity_firm
from pystoned.constant import CET_MULT, FUN_COST, RTS_VRS, RED_MOM
```

```python
# import Finnish electricity distribution firms data
data = load_Finnish_electricity_firm(x_select=['Energy', 'Length', 'Customers'],
                                    y_select=['TOTEX'])

# build and optimize the CNLS model
model = CNLS.CNLS(data.y, data.x, z=None, cet=CET_MULT, fun=FUN_COST, rts=RTS_VRS)
model.optimize('patrickfrancoalves@yahoo.com.br')
```

```python
# calculate and print unconditional expected inefficiency (mu)
rd = StoNED.StoNED(model)

print(rd.get_unconditional_expected_inefficiency(RED_MOM))
```

```python
# Import necessary libraries
import numpy as np

# Function to calculate RAROC
def calculate_raroc(expected_return, economic_capital, risk_free_rate):
    """
    Calculate the RAROC index.
    
    Parameters:
    expected_return (float): The expected return of the investment or business unit.
    economic_capital (float): The economic capital allocated to the investment or business unit.
    risk_free_rate (float): The risk-free rate of return.
    
    Returns:
    float: The RAROC index.
    """
    # Calculate the risk-adjusted return
    risk_adjusted_return = expected_return - risk_free_rate
    
    # Calculate RAROC
    raroc = risk_adjusted_return / economic_capital
    
    return raroc

# Example usage
expected_return = 0.15  # 15% expected return
economic_capital = 0.10  # 10% economic capital
risk_free_rate = 0.03  # 3% risk-free rate

raroc_index = calculate_raroc(expected_return, economic_capital, risk_free_rate)
print(f"The RAROC index is: {raroc_index:.2f}")
```
