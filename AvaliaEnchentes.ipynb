## 1.1) Leitura dos Dados de Exemplo

```python
# Importando bibliotecas necessárias
import pandas as pd


#Importando o conjunto de dados
covid = pd.read_csv(
    filepath_or_buffer = './dados/analise_ada_rhguaiba_municipios_10052024.csv', 
    sep=',',
    decimal='.',
    dtype = {'CD_MUN' : 'str' }
)

covid.rename(columns = {'CD_MUN':'id'},inplace=True)
```

## 1.2) Leitura das Coordenadas

```python
import json

# Abra o arquivo JSON e carregue os dados
with open('geojs-100-mun.json', 'r', encoding='utf-8') as f:
    geo_cidades = json.load(f)
```

```python
#geo_cidades['features'][1]
```

```python
# transformando o 'id' para tipo object, o mesmo do geoespacial
covid['id'] = covid['id'].astype('str')
```

<a name="mapaexemplo"></a>

## 1.3) Percentual de Pessoas Afetadas


```python
# Importando biblioteca necessaria
import plotly.express as px
```

```python
# Criando o mapa
mapa_agencias = px.choropleth_mapbox(
    data_frame = covid, 
    geojson = geo_cidades, 
    locations='id', 
    featureidkey='properties.id',
    color='Pessoas_%',
    color_continuous_scale= 'reds',
    range_color=(0, 1),
    mapbox_style='open-street-map',
    zoom=5, 
    #center = {"lat": -17.14, "lon": -57.33},
    center = {"lat": -30.033056, "lon": -51.230000},
    opacity=1,
    labels={'Financiamentos em Andamento':'Tx.  (/ R$ hab.)', 
            'id' : 'Código do município'
    },
    width = 1000,
    height = 800,
    title = 'Percentual de Pessoas Afetadas por municípios, 11/08/2022'
)

# Ajustando as margens
#mapa_agencias.update_layout(margin={'r':0,'t':0,'l':0, 'b':0})

# Reduzindo a largura das bordas dos municípios
#mapa_agencias.update_traces(marker_line_width=0.01)

# Exibindo o mapa
#mapa_agencias.show()
#mapa1
```
